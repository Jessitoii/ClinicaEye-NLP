### 5. Deployment Packaging
Moving final artifacts to Colab-specific model storage.

```python
import shutil

COLAB_MODEL_DIR = PROJECT_ROOT / "ai" / "models" / "Colab"
FINAL_MODEL_DIR = PROJECT_ROOT / "ai" / "models" / "final_model"

COLAB_MODEL_DIR.mkdir(parents=True, exist_ok=True)

print(f"Moving final model artifacts to {COLAB_MODEL_DIR}...")

if (FINAL_MODEL_DIR / "onnx").exists():
    # Copy ONNX and PyTorch outputs
    if (COLAB_MODEL_DIR / "onnx").exists():
        shutil.rmtree(COLAB_MODEL_DIR / "onnx")
    shutil.copytree(FINAL_MODEL_DIR / "onnx", COLAB_MODEL_DIR / "onnx")
    
    if (COLAB_MODEL_DIR / "pytorch").exists():
        shutil.rmtree(COLAB_MODEL_DIR / "pytorch")
    shutil.copytree(FINAL_MODEL_DIR / "pytorch", COLAB_MODEL_DIR / "pytorch")
    
    print("[✓] Deployment artifacts moved successfully.")
else:
    print("[!] Failed to find exported ONNX model at predictable location.")
```
